## 1. Методы Pandas

В данном модуле мы разберем базовые конструкции библиотеки Pandas с которой начинается любой анализ данных

````python
import numpy as np
import pandas as pd

pd.set_option("display.precision", 2)
````

Используем функцию read_csv, чтобы открыть датасет в Pandas. Далее используем функцию head, чтобы посмотреть первые 5 результатов:

````python
df = pd.read_csv("telecom_churn.csv")
df.head()
````

Каждая строка представляет собой одного клиента, а колонки - описательную статистику по клиенту

Далее обратим внимание на размерность данных и их особенности

````python
print(df.shape)
````

Как видим, у нас есть 3333 строки и 20 столбцов в наших данных

````python
print(df.columns)
````

Далее посмотрим основную информацию о типах данных, которые используются в нашем датасете

````python
print(df.info())
````

`bool`, `int64`, `float64`, `object` наши типы данных. Есть логическия переменная (`bool`), 3 переменных `object` и 16 числовых переменных. Важно обратить внимание, что методы машинного обучения, преимущественно работают только с числовыми данными, поэтому постараемся привести все нечисловые переменные к этому виду.

Мы можем **поменять тип данных** с помощью метода `astype`. Применим его к `Churn` :

````python
df["Churn"] = df["Churn"].astype("int64")
````

`describe` дает нам описательную статистику всех числовых колонок: пропущенные значения, среднее, стандартное отклоениение, квартили и пр.

````python
df.describe()
````

Применим метод value_counts, чтобы оценить нашу целевую переменную:

````python
df["Churn"].value_counts()
````

````python
df["Churn"].value_counts(normalize=True)
````


### Сортировка

Можно отсортировать датасет по желаемомоу признаку:


````python
df.sort_values(by="Total day charge", ascending=False).head()
````

Это можно сделать и по нескольким колонкам:

````python
df.sort_values(by=["Churn", "Total day charge"], ascending=[True, False]).head()
````

Далее познакомимся с важнейшей функцией Pandas, которая обеспечивает возможность получения необходимых нам выборок в данных.

Синтаксис такой: `df[P(df['Name'])]`, где `P` — это некоторое логическое условие, которое проверяется для каждого элемента столбца `Name`. Результатом такой индексации является DataFrame, состоящий только из строк, удовлетворяющих условию `P` в столбце `Name`.

Давайте воспользуемся им, чтобы ответить на вопрос:

**Каковы средние значения числовых характеристик для ушедших пользователей?**

````python
df[df["Churn"] == 1].mean()
````

**Сколько времени (в среднем) ушедшие пользователи проводят в телефоне в дневное время?**

````python
df[df["Churn"] == 1]["Total day minutes"].mean()
````


### Применение функций к ячейкам, столбцам и строкам

**Чтобы применить функции к каждому столбцу, используйте `apply()`:**


````python
df.apply(np.max)
````

Метод `apply` также можно использовать для применения функции к каждой строке. Для этого укажите `axis=1`. Лямбда-функции очень удобны в таких сценариях. Например, если нам нужно выбрать все состояния, начинающиеся с W, мы можем сделать это так:

````python
df[df["State"].apply(lambda state: state[0] == "W")].head()
````

Метод `map` можно использовать для **замены значений в столбце**, передав в качестве аргумента словарь вида `{old_value: new_value}`:


````python
d = {"No": 0, "Yes": 1}
df["International plan"] = df["International plan"].map(d)
df.head()
````

````python
df = df.replace({"Voice mail plan": d})
df.head()
````


### Группировка

В общем случае группировка данных в Pandas работает следующим образом:



```python
df.groupby(by=grouping_columns)[columns_to_show].function()
```

Вот пример, где мы группируем данные в соответствии со значениями переменной `Churn` и отображаем статистику по трем столбцам в каждой группе:

````python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
````

Давайте сделаем то же самое, но немного иначе, передав список функций в  `agg()`:

````python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
````


### Сводные таблицы

Предположим, мы хотим посмотреть, как распределяются наблюдения в нашей выборке в разрезе двух переменных — `Churn` и `International plan`. Для этого мы можем использовать метод перекрестной таблицы:


````python
pd.crosstab(df["Churn"], df["International plan"])
````

````python
pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
````

Давайте посмотрим на среднее количество дневных, вечерних и ночных звонков по коду города:

````python
df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
````


### Преобразование датасета

Как и многие другие вещи в Pandas, добавление столбцов в DataFrame возможно разными способами.


````python
total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)
df.insert(loc=len(df.columns), column="Total calls", value=total_calls)
df.head()
````

Есть метод и проще, без создания промежуточного экземпляра Series:

````python
df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
````

Пример метода удаления столбцов:

````python
df.drop(["Total charge", "Total calls"], axis=1, inplace=True)
````

## 2. Первая попытка предсказать отток


Давайте посмотрим, как отток связан с *International plan*. Мы сделаем это с помощью перекрестной таблицы, а также с помощью визуального анализа с помощью Seaborn. 

````python
pd.crosstab(df["Churn"], df["International plan"], margins=True)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
sns.countplot(x="International plan", hue="Churn", data=df);
````

Мы видим, что с *International Plan* уровень оттока намного выше, что является интересным наблюдением! Возможно, большие и плохо контролируемые расходы на международные звонки очень конфликтны и приводят к недовольству клиентов оператора связи.

Далее, давайте рассмотрим еще одну важную колонку – *Customer service calls*. Также составим сводную таблицу и картинку.

````python
pd.crosstab(df["Churn"], df["Customer service calls"], margins=True)
````

````python
sns.countplot(x="Customer service calls", hue="Churn", data=df);
````

Хотя это не так очевидно из сводной таблицы, на приведенном выше графике легко увидеть, что скорость оттока клиентов резко возрастает после 4 обращений в службу поддержки клиентов и выше.

Теперь давайте добавим бинарную функцию в наш DataFrame — `Customer service calls > 3`. И еще раз, давайте посмотрим, как это связано с оттоком.

````python
df["Many_service_calls"] = (df["Customer service calls"] > 3).astype("int")

pd.crosstab(df["Many_service_calls"], df["Churn"], margins=True)
````

````python
sns.countplot(x="Many_service_calls", hue="Churn", data=df);
````

Таким образом, мы видим, что повышенное количество звонков в службу поддержки и наличие международного плана склоняют пользователей отказаться от услуг компании. И эту информацию мы получили еще не применив ни одного алгоритма машинного обучения, исключительно используя функции пандас и средств визуализации!
